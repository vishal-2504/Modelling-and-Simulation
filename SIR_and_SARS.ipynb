```
import numpy as np
import scipy as sp
import pandas as pd
import math
import matplotlib.pyplot as plt
```

```
#SIR model of influenza
#no deaths and no new births, obtaining plot  6.2.3
#total 763 boys in the hostel

i0=1   #initial infected
a=0.5  #recovery rate
r0=0   #initial recovered
s0=762  #initial susceptible
ir=0.00218  #infection rate
dt=0.2
a=a*dt
ir=ir*dt

t=np.arange(0,16,0.2)

def func(a,ir):
    s=[]
    i=[]
    r=[]
    s.append(s0)
    i.append(i0)
    r.append(r0)
    for it in range(len(t)-1):
      x=(s[-1]*i[-1]*ir)
      y=x-(a*i[-1])
      z=a*i[-1]
      s.append(s[-1]-x)
      i.append(i[-1]+y)
      r.append(r[-1]+z)
    return np.array(s),np.array(i),np.array(r)

s,i,r=func(a,ir)
plt.plot(t,s)
plt.plot(t,i)
plt.plot(t,r)
plt.legend(['Susceptible','Infected','Recovered'])
plt.title('SIR model of influenza')
plt.xlabel('time')
plt.ylabel('Population')
plt.grid()
plt.show()
```

```
#Reproductive number R

i0=1   #initial infected
a=0.5  #recovery rate
r0=0   #initial recovered
s0=762  #initial susceptible
ir=0.00218  #infection rate

t=np.arange(0,10,1)

def func(a,ir,R):
    s=[]
    i=[]
    r=[]
    s.append(s0)
    i.append(i0)
    r.append(r0)
    for it in range(len(t)-1):
      x=i[-1]*R
      y=x-(a*i[-1])
      z=a*i[-1]
      s.append(s[-1]-x)
      i.append(i[-1]+y)
      r.append(r[-1]+z)
    return np.array(i)

i1=func(a,ir,0.8)
i2=func(a,ir,0.9)
#i3=func(a,ir,0.7)
i4=func(a,ir,1)
#i3=func(a,ir,3)

plt.plot(t,i1)
plt.plot(t,i2)
#plt.plot(t,i3)
plt.plot(t,i4)
plt.legend(['r=0.8','r=0.9','r=1'])
plt.title('Reproductive number Impacts')
plt.xlabel('time')
plt.ylabel('infected population')
plt.grid()
plt.show()
```

```
#Role of lockdown, assuming contacts decrease by factor of 5 in lockdown and original value after lockdown 
#SIR model of influenza
#lockdown from day 4-7
#total 763 boys in the hostel

i0=1   #initial infected
a=0.5  #recovery rate
r0=0   #initial recovered
s0=762  #initial susceptible
ir=0.00218  #infection rate
dt=0.2
a=a*dt
ir=ir*dt

t=np.arange(0,30,0.2)

def func(a,ir,d1,d2):
    s=[]
    i=[]
    r=[]
    s.append(s0)
    i.append(i0)
    r.append(r0)
    for it in range(len(t)-1):
      if t[it]<=d1 and t[it]>=d2:
        temp=ir
        ir=ir/5
        x=(s[-1]*i[-1]*ir)
        y=x-(a*i[-1])
        z=a*i[-1]
        s.append(s[-1]-x)
        i.append(i[-1]+y)
        r.append(r[-1]+z)
        ir=temp
      else:
        x=(s[-1]*i[-1]*ir)
        y=x-(a*i[-1])
        z=a*i[-1]
        s.append(s[-1]-x)
        i.append(i[-1]+y)
        r.append(r[-1]+z)
    return np.array(s),np.array(i),np.array(r)

s,i,r=func(a,ir,7,4)

plt.plot(t,s)
plt.plot(t,i)
plt.plot(t,r)
plt.legend(['Susceptible','Infected','Recovered'])
plt.title('lockdown from day 4-7')
plt.xlabel('time')
plt.ylabel('Population')
plt.grid()
plt.show()

s,i,r=func(a,ir,15,4)
#print(f)
plt.plot(t,s)
plt.plot(t,i)
plt.plot(t,r)
plt.legend(['Susceptible','Infected','Recovered'])
plt.title('lockdown from day 4-15')
plt.xlabel('time')
plt.ylabel('Population')
plt.grid()
plt.show()
```

```
#Role of lockdown Taking exponentially contacts
#SIR model of influenza
#lockdown from day 4-7
#total 763 boys in the hostel

i0=1   #initial infected
a=0.5  #recovery rate
r0=0   #initial recovered
s0=762  #initial susceptible
ir=0.00218  #infection rate
dt=0.2
a=a*dt
ir=ir*dt

t=np.arange(0,30,0.2)

def func(a,ir,d1,d2):
    s=[]
    i=[]
    r=[]
    s.append(s0)
    i.append(i0)
    r.append(r0)
    ir_arr=np.zeros(len(t))
    arr=[]
    arr.append(ir)
    x=np.log(5)/5
    t2=np.arange(0,5,0.2)
    for it in t2:
      arr.append(arr[-1]-(x*dt*arr[-1]))
    t2=np.arange(0.2,5,0.2)
    for it in t2:
      arr.append(arr[-1]+(x*dt*arr[-1]))
    j=0
    for it in range(len(t)-1):
      if t[it]<=d1+5 and t[it]>=d2:
        ir_arr[it]=arr[j]
        j+=1
      else:
        ir_arr[it]=ir

    for it in range(len(t)-1):
        ir=ir_arr[it]
        x=(s[-1]*i[-1]*ir)
        y=x-(a*i[-1])
        z=a*i[-1]
        s.append(s[-1]-x)
        i.append(i[-1]+y)
        r.append(r[-1]+z)
    return np.array(s),np.array(i),np.array(r),np.array(ir_arr)

s,i,r,ir_arr=func(a,ir,8,4)
plt.plot(t,ir_arr)
plt.title('Infection rate values(lockdown 4-8)')
plt.xlabel('time')
plt.ylabel('Infection rate')
plt.grid()
plt.show()
plt.figure()
plt.plot(t,s)
plt.plot(t,i)
plt.plot(t,r)
plt.legend(['Susceptible','Infected','Recovered'])
plt.title('lockdown from day 4-8')
plt.xlabel('time')
plt.ylabel('Population')
plt.grid()
plt.show()
```
