```
import numpy as np
import scipy as sp
import pandas as pd
import math
import matplotlib.pyplot as plt
```

```
#external influence model
c=1000
t=np.arange(0,160,1)

def func(p,q):

    n=[0]
    n1=[]
    for i in range(len(t)-1):
        n1.append((p+(q*n[i]/c))*(c-n[i]))
        n.append(n[i]+n1[i])
    n1.append(n1[-1])
    return np.array(n),np.array(n1)

p1,p2=func(0.03,0)
#p2=func(0.2,0)
#p3=func(0.3,0)
p1=(p1/c)*100
fig = plt.figure()
plt.xlabel('time in days')
ax1 = fig.add_subplot(111)
ax1.plot(t, p1)
ax1.set_ylabel('Market Share %')

ax2 = ax1.twinx()
ax2.plot(t, p2, 'r-')
ax2.set_ylabel('Consumers', color='r')
for tl in ax2.get_yticklabels():
    tl.set_color('r')
plt.title('External Influence only p=0.03')
```

```
#Mixed Influence (Bass)
# q according to 2nd qustion 
#b as Bita
c=1000
t=np.arange(0,40,1)

def func(p,q,b):

    n=[0]
    n1=[]
    for i in range(len(t)-1):
        n1.append((p+(q*pow(n[i]/c,b+1)))*(c-n[i]))
        n.append(n[i]+n1[i])
    n1.append(n1[-1])
    return np.array(n),np.array(n1)

p1,p2=func(0.03,0.2,1)
#p2=func(0.2,0)
#p3=func(0.3,0)
p1=(p1/c)*100
plt.plot(t,p1)
plt.plot(t,p2)
plt.legend(['Market Share %','Consumers Adopting'])
plt.xlabel('time in days')
plt.ylabel('Market share%')
plt.title('Internal Influence only p=0.03 & q=0.38 & beta=1')

plt.figure()
p1,p2=func(0.03,0.2,2)
#p2=func(0.2,0)
#p3=func(0.3,0)
p1=(p1/c)*100
plt.plot(t,p1)
plt.plot(t,p2)
plt.legend(['Market Share %','Consumers Adopting'])
plt.xlabel('time in days')
plt.ylabel('Market share%')
plt.title('Internal Influence only p=0.03 & q=0.38 & beta=2')

plt.figure()
p1,p2=func(0.03,0.2,3)
#p2=func(0.2,0)
#p3=func(0.3,0)
p1=(p1/c)*100
plt.plot(t,p1)
plt.plot(t,p2)
plt.legend(['Market Share %','Consumers Adopting'])
plt.xlabel('time in days')
plt.ylabel('Market share%')
plt.title('Internal Influence only p=0.03 & q=0.38 & beta=3')
```
