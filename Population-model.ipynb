```
import numpy as np
import scipy as sp
import pandas as pd
import math
import matplotlib.pyplot as plt
```

```
#Plotting f versus x
#k is stable fixed point
k=1000
x2=np.linspace(0,1500,15001,endpoint=True)
x=1-(x2/k)
x=np.multiply(x,x2)
#print(len(x))
plt.plot(x2,x)
plt.title('f versus x')
plt.ylabel('f(x)')
plt.xlabel('x')


#>0 implies unstable and <0 implies stable
plt.figure()
x=1-((2*x2)/k)
plt.plot(x2,x)
plt.title('df/dx versus x')
plt.ylabel('df/dx')
plt.xlabel('x')
```

```
#Without harvesting
#3 different cases for k=1000
#1)n0=100 
#2)n0=1500
#3)n0=600

r=0.5   #birth rate
t=np.linspace(0,20,201, endpoint = True)  #time in years
tw=t*r          #tau
delta=0.1*r
h=0        #harvesting rate
k=1000      #capacity
h2=h/(r*k)   #h_tilda
n0=100        #initial population
n2=[]
n2.append(n0/k)
for i in tw:
    y=n2[-1]*(1-n2[-1])
    y=y-h2
    y=y*delta
    y=y+n2[-1]
    n2.append(y)
n2.pop()
n=[]
for i in n2:
  n.append(i*k)
plt.plot(t,n)


n0=1500       #initial population
n2=[]
n2.append(n0/k)
for i in tw:
    y=n2[-1]*(1-n2[-1])
    y=y-h2
    y=y*delta
    y=y+n2[-1]
    n2.append(y)
n2.pop()
n=[]
for i in n2:
  n.append(i*k)
#plt.figure()
plt.plot(t,n)



n0=600         #initial population
n2=[]
n2.append(n0/k)
for i in tw:
    y=n2[-1]*(1-n2[-1])
    y=y-h2
    y=y*delta
    y=y+n2[-1]
    n2.append(y)
n2.pop()
n=[]
for i in n2:
  n.append(i*k)
#plt.figure();
plt.plot(t,n)
plt.xlabel('time in years')
plt.ylabel('Population of a specie')
plt.title('r=0.5')
plt.legend(["n0=100", "n0=1500","n0=600"])
```
